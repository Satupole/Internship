```python
import selenium
import pandas as pd
import time

from selenium import webdriver

from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException

import re
from selenium.webdriver.common.by import By

import warnings
warnings.filterwarnings('ignore')
```

```python
#connecting to driver
driver=webdriver.Chrome(r"D:\softwares\chromedriver_win32\chromedriver.exe")

#maximizing browser window
driver.maximize_window()
```

```python
#opening cars24.com website
url='https://www.cars24.com/'
driver.get(url)
time.sleep(5)
```

```python
#selecting location
manual_loc=driver.find_element(By.XPATH,'/html/body/div[1]/div[8]/div/div/div/div[2]/div[2]/button')
manual_loc.click()
```

```python
del_loc=driver.find_element(By.XPATH,'/html/body/div[1]/div[8]/div/div/div/div/div[2]/ul/li[1]')
del_loc.click()
```

```python
#selecting all cars option
all_cars_opt=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div[2]/div/div[2]/button[1]')
all_cars_opt.click()
```

```python
brand=[]
model=[]
year=[]
try:
    description=driver.find_elements(By.XPATH,'//h3[@class="_3FpCg"]')
    for i in description:
        brand.append(i.text.split()[1])
        year.append(i.text.split()[0])
        model.append(i.text.split()[2:])
except NoSuchElementException:
    brand.append('-')
```

```python
print(len(brand))
print(len(model))
print(len(year))
```

```python
transmission=[]
try:
    description=driver.find_elements(By.XPATH,'//p[@class="cvakB"]/span[1]')
    for i in description:
        transmission.append(i.text)
        
except NoSuchElementException:
    transmission.append('-')
```

```python
print(len(transmission))
```

```python
km=[]

try:
    description=driver.find_elements(By.XPATH,'//ul[@class="bVR0c"]/li[1]')
    for i in description:
        km.append(i.text)
            
except NoSuchElementException:
    km.append('-')
```

```python
print(len(km))
```

```python
owner=[]
try:
    description=driver.find_elements(By.XPATH,'//ul[@class="bVR0c"]/li[2]')
    for i in description:
        owner.append(i.text)
            
except NoSuchElementException:
    owner.append('-')
```

```python
print(len(owner))
```

```python
fuel=[]
try:
    description=driver.find_elements(By.XPATH,'//ul[@class="bVR0c"]/li[3]')
    for i in description:
        fuel.append(i.text)
            
except NoSuchElementException:
    fuel.append('-')
```

```python
print(len(fuel))
```

```python
price=[]
try:
    description=driver.find_elements(By.XPATH,'//div[@class="_7udZZ"]')
    for i in description:
        price.append(i.text)
            
except NoSuchElementException:
    price.append('-')
```

```python
print(len(price))
```

```python
location=[]
try:
    description=driver.find_elements(By.XPATH,'//ul[@class="_3oLON"]/li[3]')
    for i in description:
        location.append(i.text)
            
except NoSuchElementException:
    location.append('-')
```

```python
print(location)
del_location=location*1000
len(del_location)
```

```python
delhi_data=pd.DataFrame({'brand':brand,'model':model,'year':year,'transmission':transmission,'kms_driven':km,
                        'owner':owner,'fuel_type':fuel,'price':price,'location':del_location})
delhi_data
```
