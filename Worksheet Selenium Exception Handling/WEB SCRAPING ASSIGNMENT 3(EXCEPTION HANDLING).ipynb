```python
#opening hostel world  website
url='https://www.hostelworld.com/'
driver.get(url)
time.sleep(5)
```

```python
#searching london location in search bar
search_btn=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[1]/div/div/div[4]/div/div[2]/div/div[1]/div/div/div/input')
search_btn.send_keys('London')
```

```python
#scraping Hostel names
Name=[]
try:
    Name_tag=driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
    for j in Name_tag:
        Name.append(j.text)
except NoSuchElementException:
    Name.append('-')
```

```python
#scraping hostel distance from city
distance=[]
try:
    distance_tag=driver.find_elements(By.XPATH,'//div[@class="subtitle body-3"]')
    for j in distance_tag:
        distance.append(j.text)
except NoSuchElementException:
    distance.append('-')
```

```python
#scraping hostel ratings
ratings=[]
try:
    ratings_tag=driver.find_elements(By.XPATH,'//div[@class="score orange big"]')
    for j in ratings_tag:
        ratings.append(j.text)
except NoSuchElementException:
    ratings.append('-')
```

```python
#scraping for hostel total reveiews
reviews=[]
try:
    reviews_tag=driver.find_elements(By.XPATH,'//div[@class="reviews"]')
    for j in reviews_tag:
        reviews.append(j.text)
except NoSuchElementException:
    reviews.append('-')
```

```python
#scraping overall review
overall_reviews=[]
try:
    overall_reviews_tag=driver.find_elements(By.XPATH,'//div[@class="keyword"]')
    for j in overall_reviews_tag:
        overall_reviews.append(j.text)
except NoSuchElementException:
    overall_reviews.append('-')
```

```python
#scraping private room prices
private_prices=[]

try:
    private_dorm_prices_tag=driver.find_elements(By.XPATH,'//div[@class="price-col"]')
    count=0
    for j in private_dorm_prices_tag:
        if count%2==0:
            private_prices.append(j.text)
        count+=1        
        
except NoSuchElementException:
    private_prices.append('-')

        
```

```python
#scraping dormitory prices
dorm_prices=[]

try:
    dorm_prices_tag=driver.find_elements(By.XPATH,'//div[@class="price-col"]')
    count=0
    for j in dorm_prices_tag:
        if count%2==1:
            dorm_prices.append(j.text)
        count+=1        
        
except NoSuchElementException:
    dorm_prices.append('-')
```

```python
#scraping for facilities available in hostel
facilities=[]
try:
    facilities_tag=driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]')
    for j in facilities_tag:
        facilities.append(j.text)
except NoSuchElementException:
    facilities.append('-')
```

```python
#scraping url's to find hostel description
url=[]
url_tag=driver.find_elements(By.XPATH,'//h2[@class="title title-6"]//a')
try:
    for i in url_tag:
        url.append(i.get_attribute('href'))
except NoSuchElementException:
    url.append('-')
```

```python
#scraping for brief hostel description
p_description=[]
for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        p_description_tag=driver.find_elements(By.XPATH,'//*[@id="__layout"]/div/div[2]/section/div[6]/div/div[2]/div[2]/div/div[2]')
        for j in p_description_tag:
            p_description.append(j.text)
    except NoSuchElementException:
        p_description.append('-')
```
