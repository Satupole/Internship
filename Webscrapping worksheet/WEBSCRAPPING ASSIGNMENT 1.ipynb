# Q1:  python program to display all the header tags from wikipedia.org.

```python
def wikipedia_header_tags(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    header=[]
    
    header.append(soup.find('span',class_='central-textlogo__image sprite svg-Wikipedia_wordmark').text.split('\n')[1])
    
    df=pd.DataFrame({'Header tag':header})
    print(df)
wikipedia_header_tags('https://www.wikipedia.org/')       
```

# Q2:  python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
def top_imdb100(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    name=[]
    rating=[]
    yor=[]
    for i in list(soup.find_all('h3',class_="lister-item-header")):
        name.append(i.text.split('\n')[2])
    
    for i in list(soup.find_all('div',class_='inline-block ratings-imdb-rating')):
        rating.append(i.text.split('\n')[2])

    for i in list(soup.find_all('span',class_='lister-item-year text-muted unbold')):
        yor.append(i.text)
    df=pd.DataFrame({'movie name':name,'rating':rating,'year of release':yor})
    print(df)
    
def top_imdb1001(url2):
    page=requests.get(url2)
    soup=BeautifulSoup(page.content)
    name1=[]
    rating1=[]
    yor1=[]
    for i in list(soup.find_all('h3',class_="lister-item-header")):
        name1.append(i.text.split('\n')[2])
    
    for i in list(soup.find_all('div',class_='inline-block ratings-imdb-rating')):
        rating1.append(i.text.split('\n')[2])

    for i in list(soup.find_all('span',class_='lister-item-year text-muted unbold')):
        yor1.append(i.text)
    
    df1=pd.DataFrame({'movie name':name1,'rating':rating1,'year of release':yor1})
    print(df1)
    
top_imdb100('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
top_imdb1001('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt')
```

# Q3: IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.


```python
def top_imdb100_hindi(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    name=[]
    rating=[]
    yor=[]

    for i in list(soup.find_all('td',class_="titleColumn")):
        name.append(i.text.split('\n')[2])
    
    for i in list(soup.find_all('td',class_='ratingColumn imdbRating')):
        rating.append(i.text.split('\n')[1])

    for i in list(soup.find_all('td',class_='titleColumn')):
        yor.append(i.text.split('\n')[2])
        
    df=pd.DataFrame({'movie name':name[:100],'rating':rating[:100],'year of release':yor[:100]})
    print(df)
    
    
top_imdb100_hindi('https://www.imdb.com/india/top-rated-indian-movies/')
```

# Q4)  display list of respected former presidents of India(i.e. Name , Term of office)

```python
def list_former_presidents(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    
    president_name=[]
    president_term=[]
    
    
    for i in soup.find_all('div',class_='presidentListing'):
        president_name.append(i.text.split('\n')[1])
        
    for i in soup.find_all('div',class_='presidentListing'):
        president_term.append(i.text.split('\n')[2])
    
    data=pd.DataFrame({'Former president name':president_name,'Term of office':president_term})
    print(data)
    
    
list_former_presidents('https://presidentofindia.nic.in/former-presidents.htm')
```

# Q5a: Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
def top_10_odi_men_team(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    odi_team=[]
    matches_played=[]
    points=[]
    odi_rating=[]
    for i in soup.find_all('span',class_='u-hide-phablet'):
        odi_team.append(i.text)
    matches_played.append(soup.find('td',class_='rankings-block__banner--matches').text)
    points.append(soup.find('td',class_='rankings-block__banner--points').text)
    odi_rating.append(soup.find('td',class_='rankings-block__banner--rating u-text-right').text.split('\n')[1])
    count=0
    for i in list(soup.find_all('td',class_='table-body__cell u-center-text'))[:18]:
        if count%2==0:
            matches_played.append(i.text)
        else:
            points.append(i.text) 
        count+=1
    for i in list(soup.find_all('td',class_='table-body__cell u-text-right rating'))[:9]:
        odi_rating.append(i.text)
                  
    df=pd.DataFrame({'odi_teams':odi_team[:10],
                    'matches_played':matches_played,
                    'points':points,
                    'odi_rating':odi_rating})
    print(df)
top_10_odi_men_team('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

# Q7)  scraping headline, time and news link from cnbc news site

```python
def cnbc_headline(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    headline=[]
    time=[]
    links=[]
    for i in list(soup.find_all('div',class_="RiverHeadline-headline RiverHeadline-hasThumbnail"))[:6]:
        headline.append(i.text.split(','))
    
    
    for i in list(soup.find_all('span',class_="RiverByline-datePublished"))[:10]:
        time.append(i.text.split(','))
 
    for div in soup.findAll('div',{'class':'RiverHeadline-headline RiverHeadline-hasThumbnail'})[:5]:
        a=div.findAll('a')
        for link in a:
            links.append(link.get('href'))
       
    print(len(headline),len(time),len(links))
    df=pd.DataFrame({'News headline':headline,
                     'Time of headline':time,
                     'urls':links})
    print(df)
    
cnbc_headline('https://www.cnbc.com/world/?region=world')
```

# Q8: Python program to scrape the paper title,author,published date and url of most downloaded articles from AI in last 90 days.

```python
def top_AI_papers(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    title=[]
    authors=[]
    dop=[]
    paper_link=[]
    
    for i in list(soup.find_all('h2',class_='sc-1qrq3sd-1 MKjKb sc-1nmom32-0 sc-1nmom32-1 hqhUYH ebTA-dR')):
        title.append(i.text)
    
    
    for i in list(soup.find_all('span',class_="sc-1w3fpd7-0 pgLAT")):
        authors.append(i.text)
    
    
    for i in list(soup.find_all('span',class_="sc-1thf9ly-2 bKddwo")):
        dop.append(i.text)
    
    
    for i in soup.find_all('a',class_='sc-5smygv-0 nrDZj'):
        paper_link.append(i.get('href'))

    
    
    df=pd.DataFrame({'Title names':title,
                     'author_names':authors,
                     'publication date':dop,
                     'paper url':paper_link})
    print(df)
    
        
top_AI_papers('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
```

# Q9: python program to scrape restaurant name, cuisine,location,ratings,image url from dineout.co.in

```python
def restaurants(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    name=[]
    cuisine=[]
    locations=[]
    ratings=[]
    image=[]
    for i in list(soup.find_all('a',class_='restnt-name ellipsis')):
        name.append(i.text)
    
    for i in soup.find_all('span',class_='double-line-ellipsis'):
        cuisine.append(i.text.split('|')[1])
  
    
    for i in soup.find_all('div',class_='restnt-loc ellipsis'):
        locations.append(i.text)

    
    for i in soup.find_all('div',class_='restnt-rating rating-4'):
        ratings.append(i.text)
 
    
    for i in soup.find_all('img',class_='no-img'):
        image.append(i.get('data-src').split('|'))
 
    
    df=pd.DataFrame({'Restaurant name':name,
                     'restaurant cuisine':cuisine,
                     'restaurant location':locations,
                     'restaurant ratings':ratings,
                    'restaurant image':image})
    print(df)
    
restaurants('https://www.dineout.co.in/delhi-restaurants/welcome-back')
```

# Q10: python program to scrape the rank, publication, h5 index and h5 medianof top publications from Google Scholar from

```python
def g_forms(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    rank=[]
    publication=[]
    h5_index=[]
    h5_median=[]
    
    for i in list(soup.find_all('td',class_='gsc_mvt_p')):
        rank.append(i.text)
    
    for i in soup.find_all('td',class_='gsc_mvt_t'):
        publication.append(i.text)
  
    count=0
    for i in list(soup.find_all('td',class_='gsc_mvt_n')):
        if count%2==0:
            h5_index.append(i.text)
        else:
            h5_median.append(i.text) 
        count+=1
    
    df=pd.DataFrame({'Rank':rank,
                     'Publication name':publication,
                     'h5 index':h5_index,
                     'h5 median':h5_median})
    print(df)
    
g_forms('https://scholar.google.com/citations?view_op=top_venues&hl=en')
```
